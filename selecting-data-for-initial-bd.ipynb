```python
import pandas as pd
import sqlite3 as sql
import os

from tqdm import tqdm
```

```python
AMOUNT = 100
```

```python
df = pd.read_csv('./data/posts.csv')
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df.rating = df.rating.apply(lambda it: int(it.replace('–', '-')))
```

```python
selected_df = df.sort_values('rating', ascending=False)[:AMOUNT]
```

```python
selected_df
```

```python
del selected_df['Unnamed: 0']
del selected_df['post_id']
del selected_df['date']
del selected_df['views_count']
del selected_df['comments_count']
del selected_df['bookmarks_count']
del selected_df['rating']
del selected_df['author_nickname']
```

```python
selected_df
```

```python
if os.path.exists('db.sqlite'):
    os.remove('db.sqlite')

sql_conn = sql.connect('db.sqlite')
cursor = sql_conn.cursor()

try:
    create_table = '''CREATE TABLE articles (
    id INTEGER PRIMARY KEY,
    title TEXT NOT NULL,
    text TEXT NOT NULL,
    embedding BLOB NOT NULL 
    )
    '''

    cursor.execute(create_table)

    for meta_row in tqdm(selected_df.iterrows(), total=AMOUNT):
        row = meta_row[1]

        title = row.title
        text = row.text
        
        embedding = 'UNKNOWN'

        query = f"INSERT INTO articles (title, text, embedding)  VALUES  (?, ?, ?)"
        data = (title, text, embedding)

        cursor.execute(query, data)
        
    sql_conn.commit()
finally:
    cursor.close()
    sql_conn.close()
```
